```python
#importação das bibliotecas 
import pandas as pd
import matplotlib.pyplot as plt 
plt.style.use("seaborn")
```

```python
df = pd.read_excel("dataset\AdventureWorks.xlsx")
```

```python
df.head()
```

```python
#Quantidade de linhas e colunas 
df.shape
```

```python
df.dtypes
```

```python
df["Valor Venda"].sum()
```

```python
#Sendo criado a coluna de custos 
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head(4)
```

```python
#Custo total 
round(df['custo'].sum(), 2) 
```

```python
df["lucro"] = df["Valor Venda"] -df["custo"]
```

```python
df.head()
```

```python
#Lucro total 
round(df["lucro"].sum(),2)
```

```python
df.head()
```

```python
df["tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.sample(4)
```

```python
df["tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head()
```

```python
df["tempo_envio"].dtype
```

```python
df.groupby("Marca")["tempo_envio"].mean()
```

```python
#Procurando valores ausentes 
df.isnull().sum()
```

```python
#realizando a busca de lucro por ano
df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum()
```

```python
#colocando em um formato de float mais facil de compreender
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum()
```

```python
#Resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum().reset_index()
lucro_ano
```

```python
#Total de produtos vendidos 
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Total de produtos vendidos 
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total de Produtos vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
#agrupamento pelo ano
df.groupby(df["Data Venda"].dt.year)["lucro"].sum().plot.bar(title="Lucro xAno")
plt.xlabel("Ano")
plt.ylabel("Receita");
```
