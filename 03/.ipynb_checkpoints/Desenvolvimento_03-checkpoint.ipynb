```python
#Sendo criado a coluna de custos 
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
#Custo total 
round(df['custo'].sum(), 2) 
```

```python
df["lucro"] = df["Valor Venda"] -df["custo"]
```
